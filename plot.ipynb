```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
table=pd.read_csv("totallarge_02502.csv",index_col=0)
table500_025=table[0:12]
table500_02=table[12:24]
table1000_025=table[24:36]
table1000_02=table[36:]
```
